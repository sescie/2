```python

import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/credit_risk.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Drop unnecessary columns
data_cleaned = data.drop(columns=['Unnamed: 0', 'number_of_defaults.1', 'age.1'])

# Handle missing values - dropping rows with missing critical information
data_cleaned = data_cleaned.dropna(subset=['country', 'job', 'location'])

# Convert 'disbursemet_date' to datetime format
data_cleaned['disbursemet_date'] = pd.to_datetime(data_cleaned['disbursemet_date'], errors='coerce')

# Convert 'remaining term' to numeric (assuming it is in months or years)
data_cleaned['remaining term'] = pd.to_numeric(data_cleaned['remaining term'], errors='coerce')

# Display cleaned data structure
data_cleaned.info(), data_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Set plot style for better visuals
sns.set(style="whitegrid")
# Plot distributions for loan amount, interest rate, and age
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
# Loan Amount Distribution
sns.histplot(data_cleaned['loan_amount'], kde=True, color='blue', ax=axes[0])
axes[0].set_title('Distribution of Loan Amount')
axes[0].set_xlabel('Loan Amount')
# Interest Rate Distribution
sns.histplot(data_cleaned['interest_rate'], kde=True, color='green', ax=axes[1])
axes[1].set_title('Distribution of Interest Rate')
axes[1].set_xlabel('Interest Rate')
# Age Distribution
sns.histplot(data_cleaned['age'], kde=True, color='red', ax=axes[2])
axes[2].set_title('Distribution of Age')
axes[2].set_xlabel('Age')
plt.tight_layout()
plt.show()
# Correlation Heatmap
# Select only numeric columns for correlation calculation
numeric_data = data_cleaned.select_dtypes(include=['number'])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
